```python
import numpy as np
import matplotlib.pyplot as plt
import h5py

npz_path = '/content/drive/MyDrive/try_traces.npz'

def load_np_traces(traces_path):

  data = np.load(traces_path)

  traces=data["traces"]
  #labels=data["labels"]

  return traces#, labels
```

```python
traces = load_np_traces(npz_path)
```
